```python
# init prerequisites
%reload_ext autoreload
%autoreload 2
import numpy as np
import matplotlib.pyplot as plt
import os, re, math, time, random, yaml
from pprint import pprint

print("Numpy {}".format(np.__version__))
```

```python
# generate data
def step(x):
    return 0 if x < 0 else 1

def tanh(t):
    def f(x, t=t):
        return 1 - 2.0 / (1+math.exp(t * x))
    return f

def sigmoid(t):
    def f(x, t=t):
        return 1.0 / (1+math.exp(-x / t))
    return f

def relu(x):
    return 0 if x < 0 else x
```

```python
def plot(x, y, color, title, filename):
    # setup plot
    fig, ax = plt.subplots()
    ax.plot(x,y,color)
    ax.set(
        title=title,
        ylabel='f(x)',
        xlabel='x'
    )

    plt.setp(ax.get_yticklabels(), rotation_mode="anchor", ha="right", va="center")
    plt.setp(ax.get_xticklabels(), rotation=0, rotation_mode="anchor", ha="center", va="top")

    # save and show
    svg_path = os.path.join("output", "graph", "{}.svg".format(filename))
    os.makedirs(os.path.join("output", "graph"), exist_ok=True)
    plt.savefig(svg_path, format="svg", bbox_inches='tight')
    plt.show()
    
x = np.arange(-5, 5, 0.01)
plot(x,list(map(step, x)),'m',"Step Function", "step-act")
plot(x,list(map(tanh(2), x)),'r',"Tangens Hyperbolicus (t=2)", "tanh-act-2")
plot(x,list(map(tanh(4), x)),'r',"Tangens Hyperbolicus (t=4)", "tanh-act-4")
plot(x,list(map(sigmoid(0.1), x)),'b',"Sigmoid (t=0.1)", "sigmoid-act-01")
plot(x,list(map(sigmoid(0.5), x)),'b',"Sigmoid (t=0.5)", "sigmoid-act-05")
plot(x,list(map(relu, x)),'g',"ReLu", "relu-act")
```
